```python
import tensorflow as tf
import numpy as np
import pandas as pd
from random import random
from sklearn.model_selection import train_test_split
```

```python
def generate_dataset(num_samples,test_size):
  #create a dataset
  inputs=np.array([[ random() / 2 for _ in range(2)] for _ in range(num_samples)])
  targets=np.array([[i[0] + i[1]] for i in inputs])

  x_train,x_test,y_train,y_test=train_test_split(inputs,targets,test_size=test_size)
  return x_train,x_test,y_train,y_test
```

```python
if __name__=="__main__":
 
  x_train,x_test,y_train,y_test=generate_dataset(5000,0.3)
  print("x_test length: \n {}".format(len(x_test)))
  print("y_test length: \n {}".format(len(y_test)))

  #build model: 2->5->1
  model=tf.keras.Sequential([
        tf.keras.layers.Dense(5,input_dim=2,activation="sigmoid"),
        tf.keras.layers.Dense(1,activation="sigmoid")
  ])

  #compile model
  optimiser=tf.keras.optimizers.SGD(learning_rate=0.1)
  model.compile(optimizer=optimiser,loss="MSE")

  #train model
  model.fit(x_train,y_train,epochs=100)

  #evaluate model
  print("\nModel evaluation:")
  model.evaluate(x_test,y_test,verbose=1)

  #make predictions
  data=np.array([[0.1,0.2],[0.2,0.2]])
  predictions=model.predict(data)

  print("Predicions:",predictions)

```
